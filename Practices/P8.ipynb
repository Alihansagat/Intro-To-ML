```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('credit_card.csv')
print(f"Sample of the dataset:\n{df.head()}\n")
print(f"Dimensions of the dataset: {df.shape}\n")
print(f"Summary statistics of the quantitative variables:\n{df.describe()}\n")
```

```python
print(f"Summary statistics of the qualitative variables:\n{df.describe(include='object')}\n")
```

```python
# 1. What is the Probability that the Customer has a Credit Card ?
has_card = (df['CreditCard'] == 'yes').sum()
print(f"Number of customers with a credit card: {has_card}")

total_customers = df.shape[0]
print(f"Total number of customers: {total_customers}")

probability_has_card = has_card / total_customers
print(f"Probability that a customer has a credit card: {probability_has_card}\n")
```

```python
# 2. What is the Probability that the Customer age is greater than 50 ?
age_greater_50 = (df['age'] > 50).sum()
print(f"Number of customers with age greater than 50: {age_greater_50}")

probability_age_greater_50 = age_greater_50 / total_customers
print(f"Probability that a customer has age greater than 50: {probability_age_greater_50}\n")
```

```python
# 3. What is the Probability that the Customer is illiterate ?
illiterate = (df['education'] == 'illiterate').sum()
print(f"Number of illiterate customers: {illiterate}")

probability_illiterate = illiterate / total_customers
print(f"Probability that a customer is illiterate: {probability_illiterate}\n")
```

```python
# 4. What is the Probability that the Customer has a Telephone ?
has_telephone = (df['contact'] == 'telephone').sum()
print(f"Number of customers with a telephone: {has_telephone}")

probability_has_telephone = has_telephone / total_customers
print(f"Probability that a customer has a telephone: {probability_has_telephone}\n")
```

```python
# 5. What is the Probability that the Customer is Single and has a Credit Card ?
single_with_card = ((df['marital'] == 'single') & (df['CreditCard'] == 'yes')).sum()
print(f"Number of single customers with a credit card: {single_with_card}")

probability_single_with_card = single_with_card / total_customers
print(f"Probability that a customer is single and has a credit card: {probability_single_with_card}\n")
```

```python
# 6. What is the Probability that the Customer is Male and has at most 6 Year of Education ?
male_has_education = ((df['gender'] == 'male') & ((df['education'] == 'basic.4y') | (df['education'] == 'basic.6y'))).sum()
print(f"Number of male customers with at most 6 years of education: {male_has_education}")

probability_male_has_education = male_has_education / total_customers
print(f"Probability that a customer is male and has at most 6 years of education: {probability_male_has_education}\n")
```

```python
# 7. What is the Probability that the Customer is Retired and is divorced ?
retired_divorced = ((df['job'] == 'retired') & (df['marital'] == 'divorced')).sum()
print(f"Number of retired and divorced customers: {retired_divorced}")

probability_retired_divorced = retired_divorced / total_customers
print(f"Probability that a customer is retired and divorced: {probability_retired_divorced}\n")
```

```python
# 8. Given the Customer has the credit card, find the probability that he/she is married
has_card_married = ((df['CreditCard'] == 'yes') & (df['marital'] == 'married')).sum()
print(f"Number of customers with a credit card who are married: {has_card_married}")

has_card_total = (df['CreditCard'] == 'yes').sum()
print(f"Total number of customers with a credit card: {has_card_total}")

probability_has_card_married = has_card_married / has_card_total
print(f"Probability that a customer is married given they have a credit card: {probability_has_card_married}\n")
```

```python
# 9. If a customer has a blue collar job, find the probability that he has a housing
has_job_housing = ((df['job'] == 'blue-collar') & (df['housing'] == 'yes')).sum()
print(f"Number of blue-collar customers with a housing loan: {has_job_housing}")

has_job = (df['job'] == 'blue-collar').sum()
print(f"Total number of blue-collar customers: {has_job}")

probability_has_job_housing = has_job_housing / has_job
print(f"Probability that a blue-collar customer has a housing loan: {probability_has_job_housing}\n")
```

```python
ds = pd.read_csv('kerala_floods.csv')
print(f"Sample of the dataset:\n{ds.head()}\n")
print(f"Dimensions of the dataset: {ds.shape}\n")
print(f"Number of missing values in each column:\n{ds.isnull().sum()}\n")
```

```python
# 1- We may start by preparing the required columns:
ds['june_more_than_500'] = (ds['JUN']>500).astype('int') #if more than 500, it will give one, other than that 0
ds['FLOODS_BIN'] = ds['FLOODS'].map({"YES":1,"NO":0})
ds.head()
```

```python
# 2- Select only the columns that will help to answer the main question.
data = ds[['june_more_than_500', 'FLOODS_BIN']]
print(data.head())
```

```python
# 3- Create a cross tabulation for 'FLOODS' and 'June_more_than_500'
pd.crosstab(data['FLOODS_BIN'],data['june_more_than_500'],margins=True)
```

```python
# 4- What is the probability of a rainfall more than 500mm in June?
more_than_500 = (data['june_more_than_500'] == 1).sum()
total_rows = data.shape[0]
probability_more_than_500 = more_than_500 / total_rows
print(f"Probability of a rainfall more than 500mm in June: {probability_more_than_500}\n")
```

```python
# 5- What is the probability of having a flood?
have_flood = (data['FLOODS_BIN'] == 1).sum()
probability_have_flood = have_flood / total_rows
print(f"Probability of having a flood: {probability_have_flood}\n")
```

```python
# 6- What is the probability of having a flood and a rainfall more than 500mm in June?
have_flood_and_more_than_500 = ((data['FLOODS_BIN'] == 1) & (data['june_more_than_500'] == 1)).sum()
probability_have_flood_and_more_than_500 = have_flood_and_more_than_500 / total_rows
print(f"Probability of having a flood and a rainfall more than 500mm in June: {probability_have_flood_and_more_than_500}\n")
```

```python
# 7- What is the Probability of having a flood given it rained more than 500mm in June?
have_flood_given_more_than_500 = ((data['FLOODS_BIN'] == 1) & (data['june_more_than_500'] == 1)).sum()
total_more_than_500 = (data['june_more_than_500'] == 1).sum()
probability_have_flood_given_more_than_500 = have_flood_given_more_than_500 / total_more_than_500
print(f"Probability of having a flood given it rained more than 500mm in June: {probability_have_flood_given_more_than_500}\n")
```

```python
# 8- What is the probability of a rainfall less than 500mm in June?
less_than_500 = (data['june_more_than_500'] == 0).sum()
probability_less_than_500 = less_than_500 / total_rows
print(f"Probability of a rainfall less than 500mm in June: {probability_less_than_500}\n")
```

```python
# 9- What is probability of having a flood given that it rained less than 500mm?
have_flood_given_less_than_500 = ((data['FLOODS_BIN'] == 1) & (data['june_more_than_500'] == 0)).sum()
total_less_than_500 = (data['june_more_than_500'] == 0).sum()
probability_have_flood_given_less_than_500 = have_flood_given_less_than_500 / total_less_than_500
print(f"Probability of having a flood given it rained less than 500mm in June: {probability_have_flood_given_less_than_500}\n")
```

```python
# 10- Answer this question by using Bayes Theorem and the law of the Total Probaility
P_A = (data['june_more_than_500'] == 1).sum() / data.shape[0]          # P(June > 500)
P_B = (data['FLOODS_BIN'] == 1).sum() / data.shape[0]               # P(Flood)

P_B_given_A = ((data['FLOODS_BIN'] == 1) & (data['june_more_than_500'] == 1)).sum() / (data['june_more_than_500'] == 1).sum()
P_A_given_B = (P_B_given_A * P_A) / P_B
print(f"Probability of having a rainfall more than 500mm in June given that there was a flood: {P_A_given_B}\n")
```
