# Q1. Create a Python list representing vector v = [2, -1, 3] and multiply it by scalar 4.

```python
v = [2, -1, 3]
sc =  4
v * sc
```

It will repeat (duplicate) our list 4 times, not perform multiplication as in linear algebra.
To achieve true vector multiplication, we need to create our vector using the numpy library.

```python
v_ = np.array([2, -1, 3])
v_ * sc
```

# Q2. Given vector a = [5, -3] and scalar $\lambda$ = -2

```python
a = np.array([5, -3])
lam = -2

lambda_a = lam * a

mag_a = np.linalg.norm(a)
mag_lambda_a = np.linalg.norm(lambda_a)

print("λa =", lambda_a)
print("|a| =", mag_a)
print("|λa| =", mag_lambda_a)

print("Ratio |λa| / |a| =", mag_lambda_a / mag_a)
```

# Q4 . Create vector v = [1,2,3,4] and add scalar 5.
Explain why NumPy allows this.

```python
v = np.array([[1,2,3,4]])
sc = 5
v + sc
```

NumPy allows this because of broadcasting. The scalar 5 is automatically expanded to match the shape of the vector [1, 2, 3, 4], so it behaves like [5, 5, 5, 5] and gets added element-wise.

# Q5. Given vector a = [1,2,3,4,5,6], compute its
Euclidean norm manually and using NumPy.
Hint: By "manually," I mean solving it step by step using only basic NumPy functions.

```python
a = np.array([[1,2,3,4,5,6]])

a_ = a ** 2
sum_ = np.sum(a_)
sqrt_ = np.sqrt(sum_)
print(sqrt_)

np_norm = np.linalg.norm(a)
print(np_norm)
```

# Q6. Generate a random vector of length 10 and normalize it. Print the norm before and after normalization

```python
v = np.random.rand(10)
print(v)
norm_b = np.linalg.norm(v)
print(norm_b)
v_n = v / norm_b
print(v_n)
norm_a = np.linalg.norm(v_n)
print(norm_a)
```

# Q7. Compute the dot product of vectors
a = [1,2,3,4,5,6] & b = [0,-4,3,-3,6,5]
1. Manual loop
2. NumPy np.dot() .

```python
#1
a = np.array([1,2,3,4,5,6])
b = np.array([0,-4,3,-3,6,5])

dotp_manual = 0;
for i in range(len(a)):
    dotp_manual += a[i] * b[i];

print(dotp_manual)

dotp = np.dot(a, b)
print(dotp)
```

# Q8. Verify the distributive property:
$$a · (b + c) = a · b + a · c$$
Hint: Generate three random numbers

```python
a = np.random.rand(1)
b = np.random.rand(1)
c = np.random.rand(1)
lside = np.dot(a, b+c)
rside = np.dot(a, b) + np.dot(a, c)
print(lside == rside)
```

# Q9: Compute the Euclidean distance matrix between every point in group A and every point in group B.

```python
# Group A
A = np.array([
 [ 2, 1],
 [ 5, 4],
 [ 1, 6],
 [ 7, 2]
]) 
# Group B
B = np.array([
 [0, 0],
 [6, 3],
 [3, 5]
])

print("Shape before reshaping: ",A.shape)
print("Shape before reshaping: ",B.shape)

A_ = A[:, np.newaxis, :]
B_ = B[np.newaxis, :, :]

print("Shape after reshaping: ",A_.shape)
print("Shape after reshaping: ",B_.shape)

dif = A_ - B_
print("Difference:\n" , dif)

sq_dif = dif ** 2;
print("Square difference:\n" , sq_dif)

sum_sq = np.sum(sq_dif, axis=2)
print("Shape of Square difference:\n", sum_sq.shape)
print(sum_sq)

eu_dist = np.sqrt(sum_sq)
print("Euclidean distance: ", eu_dist)
```
