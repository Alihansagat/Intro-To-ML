```python
# Question 3:
# Let the 2D vector be:
# v = np.array([4, 5])
# Do the following in one plot:
# 1. Plot the vector as an arrow or line segment from the origin (0,0) to (5,-3) .
# 2. Add dashed x- and y-axes similar to the one in the material provided (horizontal and
# vertical dashed lines through the origin).
# 3. Set the plot limits so the vector is clearly visible (choose reasonable limits).
```

```python
v = np.array([5,-3])

plt.plot([0, v[0]], [0, v[1]], label='vector')

plt.plot([0, 0], [-6, 6], linestyle='--')
plt.plot([-6, 6], [0, 0], linestyle='--')
plt.axis((-6, 6, -6, 6))
plt.legend()

```

```python
# Question 4:
# Let the 3D vector be:
# v3 = np.array([6, -5, 4])
# Do the following in one plot:
# 1. Create a 3D plot and draw the vector from the origin to (5, -4, 3) .
# 2. Label the axes as x , y , z .
```

```python
v3 = np.array([5, -4, 3])

fig = plt.figure(figsize=plt.figaspect(1))

ax = plt.axes(projection = '3d')
ax.plot([0, v3[0]],[0, v3[1]],[0, v3[2]],linewidth=3)

ax.plot([0, 0], [-5, 5], [0, 0], 'k', label='y')   # Y-axis
ax.plot([0, 0], [0, 0], [-5, 5], 'k', label='z')   # Z-axis
ax.plot([-5, 5], [0, 0], [0, 0], 'k', label='x')   # X-axis
ax.legend()
```

```python
# Question 5:
# Given:
# v1 = np.array([4, -2])
# v2 = np.array([3, 5])
# 5.1. Compute v3 = v1 + v2 and print v3
# 5.2. On the same 2D figure, plot all three** vectors v1 , v2 , and v3 from the origin.**
```

```python
v1 = np.array([4, -2])
v2 = np.array([3, 5])
v3 = v1 + v2
print(v3)

plt.plot([0, 0], [-6, 6])
plt.plot([-6, 6], [0, 0])
plt.plot([0, v1[0]],[0, v1[1]],'b',label='v1')
plt.plot([0, v2[0]],[0, v2[1]],'r',label='v2')
plt.plot([0, v3[0]],[0, v3[1]],'g',label='v1+v2')
plt.legend()
```
