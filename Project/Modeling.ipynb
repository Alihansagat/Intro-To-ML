```python
import pandas as pd

# Load the data you saved in the previous step
X_train = pd.read_csv('X_train_processed.csv')
X_test = pd.read_csv('X_test_processed.csv')
y_train = pd.read_csv('y_train_processed.csv').values.ravel() # Convert to format model expects
y_test = pd.read_csv('y_test_processed.csv').values.ravel()

print("Data loaded! Ready for modeling.")
print(f"Training shape: {X_train.shape}")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# 1. Initialize the model
baseline_model = LogisticRegression(max_iter=1000)

# 2. Train the model
baseline_model.fit(X_train, y_train)

# 3. Make predictions on the test set
y_pred_baseline = baseline_model.predict(X_test)

# 4. Show the results (Evaluation)
print("--- Baseline: Logistic Regression ---")
print(f"Accuracy: {accuracy_score(y_test, y_pred_baseline):.2f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred_baseline))
```

+ Accuracy: The percentage of patients the model guessed correctly.
+ Classification Report: This shows how well the model avoids "False Positives" (saying someone has disease when they don't) and "False Negatives" (missing a real case of disease)

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report

# 1. Инициализируем Дерево решений
tree_model = DecisionTreeClassifier(random_state=42)

# 2. Обучаем модель на тех же данных
tree_model.fit(X_train, y_train)

# 3. Делаем предсказание
y_pred_tree = tree_model.predict(X_test)

# 4. Выводим результаты
print("--- Model 2: Decision Tree ---")
print(f"Accuracy: {accuracy_score(y_test, y_pred_tree):.2f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred_tree))
```

```python
from sklearn.neighbors import KNeighborsClassifier

# 1. Инициализируем модель KNN (возьмем 5 соседей для начала)
knn_model = KNeighborsClassifier(n_neighbors=5)

# 2. Обучаем модель
knn_model.fit(X_train, y_train)

# 3. Делаем предсказание
y_pred_knn = knn_model.predict(X_test)

# 4. Выводим результаты
print("--- Model 3: KNN ---")
print(f"Accuracy: {accuracy_score(y_test, y_pred_knn):.2f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred_knn))
```

```python
from sklearn.svm import SVC

# Создаем модель SVM. Параметр C=10 и gamma='scale' обычно помогают повысить точность.
svm_model = SVC(kernel='rbf', C=10, gamma='scale', random_state=42)

# Обучаем
svm_model.fit(X_train, y_train)

# Проверяем
y_pred_svm = svm_model.predict(X_test)

print("--- Model 4: SVM (RBF Kernel) ---")
print(f"Accuracy: {accuracy_score(y_test, y_pred_svm):.2f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred_svm))
```
