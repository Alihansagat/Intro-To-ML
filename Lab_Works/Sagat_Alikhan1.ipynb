# Q8. Filter rows where Sales > 500

```python
print(market[market['Sales'] > 500])
```

# Q9. Filter sales in City = “Yangon” and Sales > 200.


```python
sales_yangon = market[(market['City'] == "Yangon") & (market['Sales'] > 200)]
print(sales_yangon)
```

# Q10. Sort all orders by Sales in descending order.


```python
market.sort_values(by = 'Sales', ascending=False)
```

# Q11. Sort by Date (ascending) and then Time (ascending).

```python
print(market.sort_values(by = ['Date', 'Time'], ascending=[True, True]))
```

# Q12. Sort by Unit price and Quantity.

```python
print(market.sort_values(by=['Unit price', 'Quantity'], ascending=[True, True]))
```

# Q13. Calculate the total sales in “Sales” per Branch.
